```python
import cv2
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Get RTSP URL from environment variable
rtsp_url = os.getenv("RTSP_URL")

# Fallback to individual components if needed
if not rtsp_url:
    username = os.getenv("RTSP_USERNAME", "admin")
    password = os.getenv("RTSP_PASSWORD", "password")
    host = os.getenv("RTSP_HOST", "192.168.1.179")
    port = os.getenv("RTSP_PORT", "554")
    rtsp_url = f"rtsp://{username}:{password}@{host}:{port}/cam/realmonitor?channel=1&subtype=0"

cap = cv2.VideoCapture(rtsp_url)

if not cap.isOpened():
    print("Error: Could not open RTSP stream.")
    exit()

ret, frame = cap.read()
```

```python
# Check if a frame was successfully read
if ret:
    # Display the captured frame
    cv2.imshow("RTSP Frame", frame)
    cv2.waitKey(0)  # Wait indefinitely until a key is pressed
    cv2.destroyAllWindows()
else:
    print("Error: Could not read frame from RTSP stream.")
```

```python
# Release the VideoCapture object
cap.release()
```
